```python
import pandas 
import numpy as np
import h5py
import os 
import wfdb
```

```python
fs = 400  # sampling frequency
lead_names = ['DI', 'DII', 'DIII', 'AVR', 'AVL', 'AVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
output_dir = "wfdb_output"
os.makedirs("wfdb_output", exist_ok=True)

with h5py.File("../data/Sami-Trop.hdf5", "r") as f:
    tracings = f['tracings'][:]

for i, signal in enumerate(tracings):
    record_name = f"record_{i:04d}"
    record_path = record_name

    # Find the non-zero region (assumes zero-padding is symmetric)
    row_sum = np.sum(np.abs(signal), axis=1)
    nonzero_indices = np.where(row_sum > 1e-6)[0]
    start = nonzero_indices[0]
    end = nonzero_indices[-1] + 1  # include last index

    trimmed_signal = signal[start:end]

    # Write to .dat and .hea
    wfdb.wrsamp(
        record_name=record_path,
        fs=fs,
        units=["mV"] * 12,
        sig_name=lead_names,
        p_signal=trimmed_signal,
        fmt=["16"] * 12
    )

    print(f"Wrote {record_name}.dat/.hea with {end - start} samples")

print("All tracings converted successfully.")
```
