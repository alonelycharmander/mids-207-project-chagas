```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid")

# Load datasets
code15 = pd.read_csv("/Users/claire/Downloads/exams (1).csv")
samitrop = pd.read_csv("/Users/claire/Downloads/exams (2).csv")

# Store them in a dictionary for easy looping
all_data = {
    'CODE-15': code15,
    'SaMi-Trop': samitrop
}

# Define the EDA function
def explore_data(df, name="Dataset"):
    print(f"\n\n===============================")
    print(f"{name} â€“ Overview")
    print("===============================")

    print(df.head())

    print("\nQuick summary:")
    print(df.describe(include='all'))

    print("\nMissing values in each column:")
    print(df.isnull().sum())

    # Age distribution
    if 'age' in df.columns:
        plt.figure()
        sns.histplot(df['age'], bins=20, kde=True)
        plt.title(f"{name} â€“ Age Distribution")
        plt.xlabel("Age")
        plt.ylabel("Number of People")
        plt.show()

    # Gender breakdown
    if 'is_male' in df.columns:
        plt.figure()
        sns.countplot(x='is_male', data=df)
        plt.title(f"{name} â€“ Gender Breakdown")
        plt.xlabel("Is Male (True/False)")
        plt.ylabel("Count")
        plt.show()

    # Normal vs abnormal ECGs
    if 'normal_ecg' in df.columns:
        plt.figure()
        sns.countplot(x='normal_ecg', data=df)
        plt.title(f"{name} â€“ Normal vs Abnormal ECGs")
        plt.xlabel("Normal ECG (True/False)")
        plt.ylabel("Count")
        plt.show()

    # Mortality
    if 'death' in df.columns:
        plt.figure()
        sns.countplot(x='death', data=df)
        plt.title(f"{name} â€“ Death Outcome")
        plt.xlabel("Death (True/False)")
        plt.ylabel("Count")
        plt.show()

    # Survival time
    if 'timey' in df.columns:
        plt.figure()
        sns.histplot(df['timey'], bins=20, kde=True)
        plt.title(f"{name} â€“ Survival Time")
        plt.xlabel("Years Since Exam")
        plt.ylabel("Count")
        plt.show()

# Run EDA for each dataset
for name, df in all_data.items():
    explore_data(df, name=name)
```
